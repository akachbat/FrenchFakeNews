```python
#supprimer les lignes qui n'ont pas de contenu
df_fakenews_cleaned = df_fakenews.dropna(axis = 0, subset=["title", "content"])
```

```python
df_fakenews_cleaned.drop_duplicates(subset=["title"], inplace=True)
```

#### Traitement et nettoyage de texte (caratères spéciaux, espace multiples, ...)

```python
def clean_text(text):
    # Remove all the special characters
    document = re.sub(r'\W', ' ', str(text))

    # remove all single characters
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)

    # Substituting multiple spaces with single space
    document = re.sub(r'\s+', ' ', document, flags=re.I)

    # Converting to Lowercase
    document = document.lower()

    return document
```

```python
for index, row in df_fakenews_cleaned.iterrows():
    df_fakenews_cleaned.at[index, 'content'] = clean_text(row.content)
    df_fakenews_cleaned.at[index, 'title'] = clean_text(row.title)    
```

```python
df_export = df_fakenews_cleaned[["domain", "URL", "title", "content"]]
#modifier les nom de colonne
df_export.columns = ["site", "url", "title", "text"]
df_export.head()
```

```python
#sauvegarde de données
df_export.to_csv("output/fake_news_cleaned.csv",index=False)
```
