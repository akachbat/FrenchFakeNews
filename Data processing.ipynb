```python
import pandas as pd
import json  
import time
from pandas.io.json import json_normalize  
```

### Récupérer les données depuis les fichiers json du dataset des articles réels téléchargés depuis WebHose

```python
df = pd.read_json(
    "D:\\Study\\Master BigData\\PFE\\french_news_dataset\\636_webhose-2016-10_20170904085710\\news_0000001.json",
    encoding="utf-8",
    orient="records"
)
```

```python
site_whitelist = [
"lemonde.fr",
"lefigaro.fr",
"linternaute.com",
"nouvelobs.com",
"20minutes.fr",
"leparisien.fr",
"lexpress.fr",
"ouest-france.fr",
"francetvinfo.fr",
"lepoint.fr",
"rtbf.be",
"lesechos.fr",
"huffingtonpost.fr",
"journaldunet.com",
"liberation.fr",
"lesoir.be",
"france24.com",
"dhnet.be",
"ladepeche.fr",
"metronews.fr",
"rfi.fr",
"rtl.fr",
"sudouest.fr",
"rtl.be",
"7sur7.be",
"europe1.fr"    
]

json_lines = list()
#50001
for i in range(1, 245308):
    #print("news_{0:07d}.json".format(i))
    with open(
        'D:\\Study\\Master BigData\\PFE\\french_news_dataset\\636_webhose-2016-10_20170904085710\\news_{0:07d}.json'.format(i), 
        encoding="utf-8"
    ) as f: 
        json_lines.append(json.load(f))

data = list()
for d in json_lines:        
    if(d["language"] == "french" and d["thread"]["site"] in site_whitelist):
        data.append({
            "uuid": d["thread"]["uuid"],
            "url": d["thread"]["url"],
            "site": d["thread"]["site"],
            "site_type": d["thread"]["site_type"],
            "section_title": d["thread"]["section_title"],
            "author": d["author"],
            "title": d["thread"]["title_full"],
            "text": d["text"],
            "language": d["language"],    
            "domain_rank": d["thread"]["domain_rank"],
            "published": d["published"],
            "crawled": d["crawled"]
})
    
df = json_normalize(data) 

df.to_csv("real_news.csv", sep='|', encoding="utf-8");

print('done')
```

## Extract only site url for verification

```python
startTime = time.time()
json_lines = list()

for i in range(1, 245308):
    #print("news_{0:07d}.json".format(i))
    with open(
        'D:\\Study\\Master BigData\\PFE\\french_news_dataset\\636_webhose-2016-10_20170904085710\\news_{0:07d}.json'.format(i), 
        encoding="utf-8"
    ) as f: 
        json_lines.append(json.load(f))

data = list()
for d in json_lines:        
    if(d["language"] == "french"):
        data.append({
           
            "url": d["thread"]["url"],
            "site": d["thread"]["site"],
            "site_type": d["thread"]["site_type"],
            "section_title": d["thread"]["section_title"],
            "domain_rank": d["thread"]["domain_rank"],
})
    
df = json_normalize(data) 

df.to_csv("real_news_sites.csv", sep=',', encoding="utf-8");

print('done')

#show time spent for processing
"{0} seconds".format(round(time.time() - startTime,2))
```
